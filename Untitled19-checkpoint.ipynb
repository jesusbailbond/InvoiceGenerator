```python
import os
from decimal import Decimal
import tkinter as tk
from tkinter import messagebox
from tkinter import filedialog
from datetime import datetime

from InvoiceGenerator.api import Invoice, Item, Client, Provider, Creator
from InvoiceGenerator.pdf import SimpleInvoice
```

```python
root = tk.Tk()

root.title("Jordan - Invoice Generator")
root.geometry("550x760")
root.resizable(width=False, height=False)
root.configure(background="#DB9BF6")
```

```python
def generate_invoice():
    if invoice_num_entry.get() == "" or customer_name_entry.get() == "":
        messagebox.showerror("Error", "Invoice number and customer name are required")
        return
    
    invoice_date = invoice_date_entry.get() if invoice_date_entry.get() != "" else "N/A"
    email = email_entry.get() if email_entry.get() != "" else "N/A"
    phone = phone_entry.get() if phone_entry.get() != "" else "N/A"
    address = address_entry.get("1.0", 'end-1c') if address_entry.get("1.0", 'end-1c') != "" else "N/A"
    product_name = product_name_entry.get() if product_name_entry.get() != "" else "N/A"
    description = description_entry.get("1.0", 'end-1c') if description_entry.get("1.0", 'end-1c') != "" else "N/A"
    price = price_entry.get() if price_entry.get() != "" else "0"
    tax = tax_entry.get() if tax_entry.get() != "" else "0"
    
    os.environ["INVOICE_LANG"] = "en"
    
    client = Client(
    summary=customer_name_entry.get(),
    address=address_entry.get("1.0", 'end-1c') if address_entry.get("1.0", 'end-1c') != "" else "N/A",
    zip_code='',
    city='',
    country='Canada',
    phone=phone_entry.get(),
    email=email_entry.get() if email_entry.get() != "" else "N/A"
)
                  
    provider = Provider(summary='Jordan Rousseau', address='', zip_code='', city='', country='', bank_name='', bank_code='', bank_account='', phone='',email='',logo_filename='logo.png')
    
    creator = Creator('Jordan', stamp_filename='stamp.png')
    
    invoice = SimpleInvoice(client, provider, creator)
    invoice.currency_locale = ''
    invoice.currency = 'CAD'
    invoice.number = invoice_num_entry.get()
    invoice.date = datetime.now()
    invoice.use_tax = True
    
    invoice.add_item(
      Item(
        count=1, 
        price=price_entry.get(), 
        description=product_name_entry.get() + " " + (description_entry.get("1.0", 'end-1c') if description_entry.get("1.0", 'end-1c') != "" else ""), 
        tax=Decimal(tax_entry.get())
    )
)
  
   

    file_name_path = filedialog.asksaveasfilename(defaultextension='.pdf')

    if file_name_path == "":
        return
    
    pdf = SimpleInvoice(Invoice)
    
    messagebox.showinfo("Success", "Invoice generated successfully")

folder_path = r"C:\Users\Jordan\Desktop\Invoice"
try:
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
    print(f"Folder '{folder_path}' is ready.")
except Exception as e:
    print(f"Error creating folder: {e}")

```

```python
title_label=tk.Label(root)
title_label.configure(background="#f8406b",font="Arial 20 bold",foreground="#ffffff",justify="center",text="Invoice Generator")
title_label.place(x=0, y=0,width=550, height=30)

invoice_num_label=tk.Label(root)
invoice_num_label.configure(background="#90ee90",font="Arial 14 bold",foreground="#333333",justify="center",text="Invoice Number")
invoice_num_label.place(x=20,y=50,width=200,height=30)

invoice_num_entry=tk.Entry(root)
invoice_num_entry.configure(background="#393d49",font="Arial 14",foreground="#90ee90",justify="center")
invoice_num_entry.place(x=250,y=50,width=280,height=30)

invoice_date_label=tk.Label(root)
invoice_date_label.configure(background="#90ee90",font="Arial 14 bold",foreground="#333333",justify="center",text="Invoice Date and Time")
invoice_date_label.place(x=20,y=100,width=200,height=30)

invoice_date_entry=tk.Entry(root)
invoice_date_entry.configure(background="#393d49",font="Arial 14",foreground="#90ee90",justify="center")
invoice_date_entry.insert(0, datetime.now())
invoice_date_entry.place(x=250,y=100,width=280,height=30)

customer_name_label=tk.Label(root)
customer_name_label.configure(background="#ffe667",font="Arial 14 bold",foreground="#333333",justify="center",text="Customer Name")
customer_name_label.place(x=20,y=150,width=200,height=30)

customer_name_entry=tk.Entry(root)
customer_name_entry.configure(background="#393d49",font="Arial 14",foreground="#ffd700",justify="center")
customer_name_entry.place(x=250,y=150,width=280,height=30)

email_label=tk.Label(root)
email_label.configure(background="#ffe667",font="Arial 14 bold",foreground="#333333",justify="center",text="Email")
email_label.place(x=20,y=250,width=200,height=30)
email_entry=tk.Entry(root)
email_entry.configure(background="#393d49",font="Arial 14",foreground="#ffd700",justify="center")
email_entry.place(x=250,y=250,width=280,height=30)


phone_label=tk.Label(root)
phone_label.configure(background="#ffe667",font="Arial 14 bold",foreground="#333333",justify="center",text="Phone")
phone_label.place(x=20,y=200,width=200,height=30)


phone_entry=tk.Entry(root)
phone_entry.configure(background="#393d49",font="Arial 14",foreground="#ffd700",justify="center")
phone_entry.place(x=250,y=200,width=280,height=30)

address_label=tk.Label(root)
address_label.configure(background="#ffe667",font="Arial 14 bold",foreground="#333333",justify="center",text="Address")
address_label.place(x=20,y=320,width=200,height=30)


address_entry=tk.Text(root)
address_entry.configure(background="#393d49",font="Arial 14",foreground="#ffd700", wrap="word")
address_entry.place(x=250,y=300,width=280,height=80)


product_name_entry=tk.Entry(root)
product_name_entry.configure(background="#393d49",font="Arial 14",foreground="#ffb367",justify="center")
product_name_entry.place(x=250,y=400,width=280,height=30)


description_label=tk.Label(root)
description_label.configure(background="#ffb367",font="Arial 14 bold",foreground="#333333",justify="center",text="")
description_label.place(x=20,y=470,width=200,height=30)


description_entry=tk.Text(root)
description_entry.configure(background="#393d49",font="Arial 14",foreground="#ffb367", wrap="word")
description_entry.place(x=250,y=450,width=280,height=80)


quantity_label=tk.Label(root)
quantity_label.configure(background="#ffb367",font="Arial 14 bold",foreground="#333333",justify="center",text="Billed Hours")
quantity_label.place(x=20,y=550,width=200,height=30)


quantity_entry=tk.Entry(root)
quantity_entry.configure(background="#393d49",font="Arial 14",foreground="#ffb367",justify="center")
quantity_entry.place(x=250,y=550,width=280,height=30)


price_label=tk.Label(root)
price_label.configure(background="#ffb367",font="Arial 14 bold",foreground="#333333",justify="center",text="Price")
price_label.place(x=20,y=600,width=200,height=30)


price_entry=tk.Entry(root)
price_entry.configure(background="#393d49",font="Arial 14",foreground="#ffb367",justify="center")
price_entry.place(x=250,y=600,width=280,height=30)


tax_label=tk.Label(root)
tax_label.configure(background="#ffb367",font="Arial 14 bold",foreground="#333333",justify="center",text="Tax (%)")
tax_label.place(x=20,y=650,width=200,height=30)


tax_entry=tk.Entry(root)
tax_entry.configure(background="#393d49",font="Arial 14",foreground="#ffb367",justify="center")
tax_entry.place(x=250,y=650,width=280,height=30)


create_invoice_button=tk.Button(root)
create_invoice_button.configure(font="Arial 14 bold",justify="center",text="Create Invoice", command=generate_invoice)
create_invoice_button.place(x=190,y=700,width=171,height=39)
```
